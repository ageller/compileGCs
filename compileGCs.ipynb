# Compile the Harris catalog into 1 csv file

and add in data from other sources as well

```python
import pandas as pd
import numpy as np
```

```python
# position data
harrisPos_df = pd.read_fwf("harris2010_pos.txt", 
                           widths = [12,12,13,13,9,9,6,6,6,6,6], 
                           names=['ID', 'otherName', 'RA2000[hr]', 'Dec2000[deg]',
                                  'L[deg]', 'B[deg]', 'R_Sun[kpc]', 'R_gc[kpc]', 
                                  'X[kpc]', 'Y[kpc]', 'Z[kpc]'],
                           skiprows = 16)
harrisPos_df
```

```python
# photometric data
harrisPhot_df = pd.read_fwf("harris2010_phot.txt", 
                           widths = [13,6,4,6,6,6,6,7,7,6,6,6,5,7], 
                           names = ['ID', '[Fe/H]', 'wt', 'E(B-V)', 'V_HB', '(m-M)V',
                                   'V_int','M_V_int', '(U-B)_int', '(B-V)_int', '(V-R)_int',
                                   '(V_I)_int','specType_int','ellipticity'],
                           skiprows = 19)
harrisPhot_df
```

```python
# velocity data
harrisVel_df = pd.read_fwf("harris2010_vel.txt", 
                           widths = [13,8,6,9,7,6,6,4,6,6,8,8,6,6], 
                           names = ['ID', 'helio_rv[km/s]', 'err_helio_rv[km/s]', 
                                    'lsr_rv[km/s]', 'sig_v0[km/s]', 'err_sig_v0[km/s]',
                                    'King_c','cc_flag','r_c[arcmin]','r_hl[arcmin]',
                                    'surface_brightness0[mag/arcsec2]',
                                    'luminosity_density0[Lsum/pc3]',
                                    'trelax(r_c)[log10(yr)]','trelax(r_h)[log10(yr)]'],
                           skiprows = 20)
harrisVel_df
```

```python
#merge these
harrisPosPhot_df = harrisPos_df.join(harrisPhot_df.set_index('ID'),on = 'ID', how = 'outer', 
                                     lsuffix='_pos', rsuffix='_phot') 

harrisPosPhotVel_df = harrisPosPhot_df.join(harrisVel_df.set_index('ID'),on = 'ID', how = 'outer', 
                                            rsuffix='_vel') 


harrisPosPhotVel_df.iloc[0]
```

## Marin-Franch 2009 data

https://ui.adsabs.harvard.edu/abs/2009ApJ...694.1498M/abstract

cat Marin-Franch2009_tab4.txt | sed 's/+or-/ /g' | sed 's/   / /g' > Marin-Franch2009_tab4_fixed.txt

```python
MarinFranch_df = pd.read_csv('Marin-Franch2009_tab4_fixed.txt', sep = ' ',skipinitialspace = True,
                           names = ['ID', '[M/H]_ZW', '[M/H]_CG', 
                                    'MtoF606W', 'errMtoF606W',
                                    'D07_ZW','errD07_ZW',
                                   'D07_CG','errD07_CG',
                                   'P04_CG','errP04_CG',
                                   'B94_CG','errB94_CG',
                                   'G00_CG','errG00_CG',
                                   'Group'],
                            skiprows = 7,
                            index_col = False)
MarinFranch_df
```

```python
norm = 12.80 #from Marin-Franch2009 Section 6.1
D07cgNorm = MarinFranch_df['D07_CG'].values*norm
print(D07cgNorm)

# use this to normalize
# age47Tuc = 11.25 ;Gyr from EB, Thompson et al. 2010 http://adsabs.harvard.edu/abs/2010AJ....139..329T
# take the mean from the Padova isochrones in their Table 9

age47Tuc = np.mean(np.array([12.39, 12.46, 12.41, 12.26, 12.34, 12.29, 11.81, 12.79]))

norm = age47Tuc/MarinFranch_df['G00_CG'].values[0]
G00cgNorm = MarinFranch_df['G00_CG'].values*norm
print(G00cgNorm)

MarinFranch_df['D07_CG_Norm'] = D07cgNorm
MarinFranch_df['G00_CG_Norm'] = G00cgNorm
```

```python
#fix the names
IDs = [x.replace('_',' ') for x in MarinFranch_df['ID'].values]
MarinFranch_df['ID'] = IDs
#MarinFranch_df
```

```python
#merge
harrisPosPhotVel_MarinFranch_df = harrisPosPhotVel_df.join(MarinFranch_df.set_index('ID'),on = 'ID', how = 'outer', 
                                            rsuffix='_MF') 


harrisPosPhotVel_MarinFranch_df.iloc[0]
```

## Milone 2012 data

https://ui.adsabs.harvard.edu/abs/2012A%26A...540A..16M/abstract

```python
Milone_df = pd.read_csv('Milone2012_tab2.readcol.txt', sep = ' ',skipinitialspace = True)
Milone_df
```

```python
#fix the names
IDs = [x.replace('_',' ') for x in Milone_df['ID'].values]
Milone_df['ID'] = IDs
```

```python
#replace the 9.999 values with NaNs
Milone_df.replace(9.999, np.nan, inplace=True)
```

```python
#merge
harrisPosPhotVel_MarinFranch_Milone_df = harrisPosPhotVel_MarinFranch_df.join(Milone_df.set_index('ID'),
                                                                               on = 'ID', how = 'outer', 
                                                                               rsuffix='_Mi') 


harrisPosPhotVel_MarinFranch_Milone_df.iloc[0]
```

```python
#add a mass column assuming a M/L = 2
MV = harrisPosPhotVel_MarinFranch_Milone_df['M_V_int']
BC = -0.15 #for MS stellar type around mid G (not sure what to use here)
MBol = MV + BC
MBolSun = 4.74
lum = 10.**(0.4*(MBolSun - MBol)) #solar luminosity
mass = lum*2.

harrisPosPhotVel_MarinFranch_Milone_df['Mcl[MSun]'] = mass

harrisPosPhotVel_MarinFranch_Milone_df
```

```python
#write to csv
harrisPosPhotVel_MarinFranch_Milone_df.to_csv('GCcompiled.csv')
```
